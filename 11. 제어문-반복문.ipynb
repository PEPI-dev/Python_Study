# 4.range()í•¨ìˆ˜
range() í•¨ìˆ˜ëŠ” íŒŒì´ì¬ì—ì„œ ìˆœì°¨ì ì¸ ì •ìˆ˜ì˜ ì‹œí€€ìŠ¤ë¥¼ ìƒì„±í•˜ëŠ” ë° ì‚¬ìš©ë˜ëŠ” ë‚´ì¥ í•¨ìˆ˜ì…ë‹ˆë‹¤. ì£¼ë¡œ for ë¬¸ê³¼ í•¨ê»˜ ì‚¬ìš©ë˜ë©°, ì¼ì •í•œ ë²”ìœ„ ë‚´ì—ì„œ ë°˜ë³µ ì‘ì—…ì„ ìˆ˜í–‰í•  ë•Œ ìœ ìš©í•˜ê²Œ í™œìš©ë©ë‹ˆë‹¤. range() í•¨ìˆ˜ì˜ ê²°ê³¼ë¡œ ìƒì„±ë˜ëŠ” ìˆœì°¨ì ì¸ ì •ìˆ˜ ì‹œí€€ìŠ¤ëŠ” ì‹¤ì œë¡œ ë©”ëª¨ë¦¬ì— ì €ì¥ë˜ì§€ ì•Šê³  í•„ìš”í•  ë•Œë§ˆë‹¤ ìƒì„±ë˜ë¯€ë¡œ, í° ë²”ìœ„ì˜ ë°˜ë³µë¬¸ì—ë„ íš¨ìœ¨ì ìœ¼ë¡œ ì‚¬ìš©í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
```

```
* start (ì„ íƒ ì‚¬í•­): ìˆœì°¨ì ì¸ ë²”ìœ„ì˜ ì‹œì‘ ê°’ì„ ì§€ì •í•©ë‹ˆë‹¤. ê¸°ë³¸ê°’ì€ 0ì…ë‹ˆë‹¤.
* stop: ìˆœì°¨ì ì¸ ë²”ìœ„ì˜ ë ê°’ì„ ì§€ì •í•©ë‹ˆë‹¤. ìƒì„±ëœ ì‹œí€€ìŠ¤ëŠ” stop ê°’ ì§ì „ê¹Œì§€ì˜ ì •ìˆ˜ë¥¼ í¬í•¨í•©ë‹ˆë‹¤.
* step (ì„ íƒ ì‚¬í•­): ìˆœì°¨ì ì¸ ê°’ì„ ì¦ê°€ì‹œí‚¤ëŠ” ê°„ê²©ì„ ì§€ì •í•©ë‹ˆë‹¤. ê¸°ë³¸ê°’ì€ 1ì…ë‹ˆë‹¤.

* range() í•¨ìˆ˜ì˜ ë°˜í™˜ê°’ì€ range ê°ì²´ë¡œ, ì´ ê°ì²´ëŠ” ì‹¤ì œ ì •ìˆ˜ ì‹œí€€ìŠ¤ë¥¼ ì €ì¥í•˜ì§€ ì•Šê³  ë²”ìœ„ ì •ë³´ë§Œ ê°€ì§€ê³  ìˆìŠµë‹ˆë‹¤. ë”°ë¼ì„œ í•„ìš”í•œ ì •ìˆ˜ ì‹œí€€ìŠ¤ëŠ” ì‹¤ì œ ì‚¬ìš© ì‹œì— ìƒì„±ë©ë‹ˆë‹¤.


````python
num = range(0, 10, 1) # numì— range(0ë¶€í„° 10ì„ ë„£ê³  1ì”© stepì„ í•œë‹¤.)

print(num) # range(0, 10), ë©”ëª¨ë¦¬ì— 0ë¶€í„° 9ê¹Œì§€ì˜ ë°ì´í„°ê°€ ì €ì¥ë˜ì–´ìˆë‹¤.
````

````python
for i in num : # ië¼ëŠ” ë³€ìˆ˜ì— numì´ë¼ëŠ” ìœ„ì— ì„ ì–¸í•œ range(0, 10, 1)ì„ ë„£ëŠ”ë‹¤. 0ë¶€í„° 9ê¹Œì§€ ë„ëŠ”ê²ƒ!
  print(i, end = '') # iì— ë“¤ì–´ê°„ numì„ end = ' 'ìœ¼ë¡œ \nì´ ì•„ë‹Œ í•œì¤„ë¡œ í‘œì¶œí•´ë¼.
````

````python
for i in range(0, 10, 1) :
  print(i, end = '') # ì´ë ‡ê²Œ ì„ ì–¸ì„ í•˜ë©´ ë©”ëª¨ë¦¬ì— ê³„ì†ì €ì¥ë˜ì§€ ì•ŠëŠ”ë‹¤. ì‚¬ìš©í›„ ë²„ë ¤ì§.
````

````python
for i in range(0, 10) : # stepê°’ì„ ì§€ì›Œë„ ë˜‘ê°™ì€ ê°’ì„ í‘œí˜„í•œë‹¤. default = 1
  print(i, end = '')
````

# 5. enumerate()í•¨ìˆ˜
enumerate() í•¨ìˆ˜ëŠ” íŒŒì´ì¬ì—ì„œ ë°˜ë³µë¬¸ì„ ì‚¬ìš©í•  ë•Œ ì¸ë±ìŠ¤ì™€ ê°’ì„ í•¨ê»˜ ê°€ì ¸ì˜¤ê¸° ìœ„í•´ ì‚¬ìš©ë˜ëŠ” ë‚´ì¥ í•¨ìˆ˜ì…ë‹ˆë‹¤. ì£¼ë¡œ for ë¬¸ê³¼ í•¨ê»˜ ì‚¬ìš©ë˜ë©°, ë°˜ë³µë¬¸ ë‚´ì—ì„œ ìˆœì„œë¥¼ ì¶”ì í•˜ë©´ì„œ ê°’ê³¼ í•´ë‹¹ ê°’ì˜ ì¸ë±ìŠ¤ë¥¼ í•¨ê»˜ ì‚¬ìš©í•  ë•Œ ìœ ìš©í•©ë‹ˆë‹¤. enumerate() í•¨ìˆ˜ëŠ” (ì¸ë±ìŠ¤, ê°’)ì˜ íŠœí”Œì„ ë°˜í™˜í•˜ëŠ” ì´í„°ë ˆì´í„°(iterator)ë¥¼ ìƒì„±í•©ë‹ˆë‹¤.


```

```
* iterable: ë°˜ë³µ ê°€ëŠ¥í•œ ê°ì²´(ë¦¬ìŠ¤íŠ¸, íŠœí”Œ, ë¬¸ìì—´ ë“±)ë¥¼ ì§€ì •í•©ë‹ˆë‹¤.
* start (ì„ íƒ ì‚¬í•­): ì¸ë±ìŠ¤ì˜ ì‹œì‘ ê°’ì„ ì§€ì •í•©ë‹ˆë‹¤. ê¸°ë³¸ê°’ì€ 0ì…ë‹ˆë‹¤.
enumerate() í•¨ìˆ˜ëŠ” (ì¸ë±ìŠ¤, ê°’) í˜•íƒœì˜ íŠœí”Œì„ ìƒì„±í•˜ì—¬ ì´í„°ë ˆì´í„°ë¥¼ ë°˜í™˜í•©ë‹ˆë‹¤. ì´ë¥¼ for ë¬¸ê³¼ í•¨ê»˜ ì‚¬ìš©í•˜ì—¬ ì¸ë±ìŠ¤ì™€ ê°’ì„ ë™ì‹œì— ê°€ì ¸ì™€ ì²˜ë¦¬í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.


````python
li1 = [10, 20, 30, 40, 50]

for e in enumerate(li1) :
  print (e)

# (0, 10) ì•ì€ indexìˆœì„œ, ë’¤ëŠ” ê°’
# (1, 20) ì´ë ‡ê²Œ tupleë¡œ ë¦¬í„´ì‹œì¼œì£¼ëŠ” ì¢‹ì€í•¨ìˆ˜.
# (2, 30)
# (3, 40)
# (4, 50)
````

````python
# ì¢€ë” í¸í•˜ê²Œ ì‚¬ìš©í•˜ëŠ”ë°©ë²•
for i, v in enumerate(li1) :
  print(f"index : {i}, value : {v}")

# index : 0, value : 10
# index : 1, value : 20
# index : 2, value : 30
# index : 3, value : 40
# index : 4, value : 50
````

