# 2. 셀의 종류
* 코드 : 파이썬 코드 및 리눅스 명령어를 실행할 수 있는 셀
* 텍스트 : 설명을 닫거나 문구 등을 적기 위한 텍스트 전용 셀

### 2-1. 셀의 삽입 방법
* 상단의 "+코드", "+텍스트" 버튼을 눌러 삽입
* 코드셀을 위에 삽입 : 컨트를 + M + A
* 코드셀을 아래에 삽입 : 컨트롤+ M + B

### 2-2. 셀을 변경하는 방법
* 코드셀을 텍스트셀로 변경 : 컨트롤 + M + M
* 텍스트셀을 코드셀로 변경 : 컨트롤 + M + Y

### 2-3. 셀을 삭제하는 방법
* 우측의 휴지통 아이콘을 클릭하여 삭제
* 단축키 : 컨트롤 + M + D

# 3. 셀의 실행
* 좌측의 실행 아이콘 클릭
* 컨트롤 + Enter : 해당 셀을 실행하고 커서는 해당 셀에 남음
* Shift + Enter : 해당 셀을 실행하고 다음셀로 넘어감
* Alt + Enter : 해당 셀을 실행하고 커서는 다음셀을 삽입 후 넘어감
