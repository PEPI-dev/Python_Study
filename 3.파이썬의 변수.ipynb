# 1. 변수
변수는 프로그래밍에서 데이터를 저장하고 관리하는 데 사용되는 중요한 개념입니다. 변수는 이름이 붙은 메모리 공간으로, 값을 저장하거나 참조할 때 사용됩니다.

```python

name = '김사과'
age = 20

print(name)
print(age)

name = '하하'
print(name)
```

# 2. 출력서식 지정과 변수


```python
num1 = 10
num2 = 3

print('%d / %d = %4.2f' % (num1,num2,num1/num2))
```

# 3. 변수의 자료형
파이썬은 동적 타이핑(dynamic typing) 언어로, 변수의 타입을 선언할 필요 없이 값을 할당할 때 자동으로 타입을 추론합니다. 이는 프로그래머가 변수의 타입을 직접 명시할 필요 없이 더 유연하게 코드를 작성할 수 있는 장점을 제공합니다. 그렇지만 이로 인해 잘못된 타입의 값이 변수에 할당되는 상황도 발생할 수 있습니다.



```python
name = 'dsfsd'
age = 20
weight = 50.5
isLover = False

print(type(name))
print(type(age))
print(type(weight))
print(type(isLover))
```

# 4. 변수 삭제
변수를 삭제하기 위해 del 문을 사용할수 있다. del문은 지정된 변수를 제거하고 해당 메모리 공간을 해제한다. 변수가 삭제되면 해당 이름으로 변수에 더 이상 접근할수 없음

```python
name1= '김사과'
name2= name1
del name1
print(name2)
```

```python
isLover = False
print(isLover)
del isLover
print(isLover)
```
