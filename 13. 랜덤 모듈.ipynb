#2. 로또번호 예상 프로그램을 만들어보자.
  * 1 ~ 45까지의 6개의 숫자를 무작위로 추출
  * 단, 숫자가 중복되면 안됨.
  * 오름차순으로 정렬.
  * 단, 보너스 숫자는 생각하지 않음.

```python
while True :
  num = int(input("숫자를 입력하세요 : ")) # 입력을 할때까지 루한루프
  if num == 6 :
    break # 6을 입력하면 while문을 빠져나감.
  else :
    print(f"{num}을 입력하였습니다.") # 계속 반복
print("프로그램을 종료합니다.")
```

```python
# 가위바위보 게임
# random 함수를 임포트
import random

# 변수 선언
msg1 = "가위, 바위, 보 게임을 시작합니다."
choice = "가위, 바위, 보 중 원하는것을 입력하세요 : "
err = "잘못된 값을 입력하셨습니다. 프로그램을 종료합니다."
victory = "🤭컴퓨터와 대전에서 승리하였습니다🤭"
defeat = "😭컴퓨터와의 대전에서 패배하였습니다😭 다시 시작합니다."
tie = "😭컴퓨터와의 대전에서 무승부하였습니다😭 다시 시작합니다."
lin20 = ("=" * 20)
vic_lin = ("=" * 30)
total_vic, total_def, total_tie = (0, 0, 0)

# 반복문 실행
while True :
  data = ["가위", "바위", "보"]
  computer_random = random.choice(data) # data의 변수에 list형식으로 지정 후 random.choice()라는 변수로 가져오도록 한다.
  print(msg1) # 변수 msg1을 선언
  user_input = input(choice)
  if user_input not in data : # 만약에 data변수에 들어간 단어가 아니면 err메세지를 표출
    print(err)
    break # 에러가 발생하면 멈춤
  if user_input == computer_random : # 만약 비겼다면?
    total_tie += 1 # 비긴함수에 +1을 카운트한다.
    print(f"{lin20}{tie}{lin20}\n                             [컴퓨터의 값] : {computer_random}, [입력된 값] : {user_input}") # 컴퓨터 및 사용자가 입력한 값을 표현
  elif (user_input == "가위" and computer_random == "바위") or (user_input == "바위" and computer_random == "보") or (user_input == "보" and computer_random == "가위") : # 만약에 졌다면?
      total_def += 1 # 진함수에  +1을 카운트한다.
      print(f"{lin20}{defeat}{lin20}\n                        [컴퓨터의 값] : {computer_random}, [입력된 값] : {user_input}") # 컴퓨터 및 사용자가 입력한 값을 표현
  else : # 만약에 위쪽 둘다 아니고 이겼다면?
    total_vic += 1 # 승리변수에 +1을 카운트한다.
    print(f"\n {vic_lin}\n {victory}\n {vic_lin}") # 승리의 메세지를 표출
    total_count = total_tie + total_def + total_vic # 카운트한 값을 total_count에 저장
    print(f"총플레이 : {total_count} | 비긴횟수 : {total_tie} | 패배횟수 : {total_def}") # 모든값을 표출
    break # 정지
```

```python
# 로또 예상 번호 프로그램
# random 함수를 임포트
import random as rd

# 변수선언
nl = []

# 반복문 실행
while True :
  rdl = rd.randrange(1, 46)
  if rdl not in nl :
    nl.append(rdl)

  if len(nl) == 6 :
    print("""    ==================
    | 6개의 숫자가 나옵니다 |
    ==================""")
    break
  nl.sort()
print(f" 1번째 숫자 : {nl[0]}\n 2번째 숫자 : {nl[1]}\n 3번째 숫자 : {nl[2]}\n 4번째 숫자 : {nl[3]}\n 5번째 숫자 : {nl[4]}\n 6번째 숫자 : {nl[5]}")
```
