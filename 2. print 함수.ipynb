# 1. print ()
print() 함수는 파이썬의 기본 출력 함수로, 터미널이나 콘솔 창에 텍스트나 데이터를 출력하기 위해 사용됩니다.

```python
print('안녕하세요 파이썬')
print(10)
print('10')
print(10, 20, 30, 40)
```

```python
print('Hello', 'Python','Hi','Python')
print('Hello',end='')
print('Python')
print('Hello',end = ' ')
print('Python')
```

# 2. 파이썬의 escape 문자
파이썬에서 이스케이프 문자는 문자열 안에서 특별한 의미를 가지는 문자들을 나타내기 위해 사용됩니다. 이스케이프 문자는 백슬래시(\)로 시작하며, 문자열 내에서 특정한 동작이나 특수 문자를 표현하기 위해 사용됩니다.

1. \n: 줄바꿈을 나타냅니다.

2. \t: 탭 문자를 나타냅니다.
3. \\: 백슬래시 자체를 나타냅니다.
4. \" 또는 \': 큰따옴표나 작은따옴표를 나타냅니다. 문자열을 감싸는 따옴표와 구분하기 위해 사용됩니다.
5. \r: 캐리지 리턴을 나타냅니다. 문자열을 출력한 후 커서를 줄의 처음으로 이동시킵니다.
6. \b: 백스페이스를 나타냅니다. 문자열 내에서 한 글자를 삭제합니다.

```python
print('Hello\Python')
print('Name:\t김사과')
print('파이썬은 너무 재밌네: \\')
print('파이썬은 너무 재밌네: \' ')
```

# 3. print 함수내 연산
print() 함수는 문자열뿐만 아니라 숫자와 연산 결과도 출력할 수 있습니다.

```python
print('더하기 : ', 10+3)
print('빼기 : ', 10 - 3)
print('곱 : ',10 * 3)
print('나누기 : ',10 / 3)
print('정수나누기 : ',10 // 3) # 지금 계산결과로 그냥 나누기하면 소숫점 까지나오는데 정수나눗셈은 몫만 나옴
print('나머지 : ',10 % 3) # 나머지
print('거듭제곱 : ',10 ** 3) #10의 3승
```

# 4. 출력서식 지정
파이썬의 print() 함수는 출력 서식을 지정하여 값을 형식화하고 출력하는 기능을 제공합니다. 서식 지정자와 함께 사용되는 % 연산자와 format() 메서드를 사용하여 원하는 형식으로 출력할 수 있습니다.

```python
print('정수: %d' % 100)
print(f'정수: {100}') # f-string
print('정수: %d' % 100.5)
print('실수: %f' % 100.5)
print('실수: %f' % 100)
print('문자열: %s' % 'Python')
print('문자열: %s' % 100)

print('5 + 3 = 8')
print('%d + %d = %d' % (5, 3, 5+3))
```

```python
print('이름 : %s, 나이 : %d' % ('김사과', 20))
print('몸무게 : %.2f' % 50.1234)
print('내친구 : {} {}'.format('오렌지', '바나나'))
```
